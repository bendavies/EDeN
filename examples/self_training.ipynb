```python
def describe(X):
    print 'Instances: %d ; Features: %d (with an avg of %d features per instance)' % (X.shape[0], X.shape[1],  X.getnnz()/X.shape[0])
```

```python
from scipy.sparse import vstack
import numpy as np
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score
from eden.util import random_bipartition_iter
from eden.util import fit_estimator

def fit_and_estimate(iterable_pos, iterable_neg, vectorizer, train_test_split=0.7, n_jobs=1, cv=10):
    #make train-test split
    iterable_pos_train, iterable_pos_test = random_bipartition_iter(iterable_pos, relative_size=train_test_split)
    iterable_neg_train, iterable_neg_test = random_bipartition_iter(iterable_neg, relative_size=train_test_split)

    X_pos_train = vectorizer.transform( iterable_pos_train, n_jobs=n_jobs )
    X_neg_train = vectorizer.transform( iterable_neg_train, n_jobs=n_jobs )
    
    #optimize hyperparameters classifier
    optpredictor = fit_estimator(positive_data_matrix=X_pos_train, negative_data_matrix=X_neg_train, cv=cv, n_jobs=n_jobs)

    X_pos_test = vectorizer.transform( iterable_pos_test, n_jobs=n_jobs )
    X_neg_test = vectorizer.transform( iterable_neg_test, n_jobs=n_jobs )
    yp =  [1] * X_pos_test.shape[0]
    yn = [-1] * X_neg_test.shape[0]
    y = np.array(yp + yn)
    X_test = vstack( [X_pos_test,X_neg_test] , format = "csr")
    print 'Test set'
    describe(X_test)
    
    print '-'*80
    print 'Test Estimate'
    predictions=optpredictor.predict(X_test)
    margins=optpredictor.decision_function(X_test)
    print classification_report(y, predictions)
    roc = roc_auc_score(y, margins)
    print 'ROC: %.3f' % roc
    apr = average_precision_score(y, margins)
    print 'APR: %.3f'% apr
    
    return optpredictor
```

```python
pos='bursi.pos.gspan'
neg='bursi.neg.gspan'
```

```python
from eden.graph import Vectorizer
vectorizer = Vectorizer( complexity=1 )
```

```python
%%time
from eden.converter.graph.gspan import gspan_to_eden
iterable_pos = gspan_to_eden( pos )
iterable_neg = gspan_to_eden( neg )
estimator = fit_and_estimate( iterable_pos, iterable_neg, vectorizer, train_test_split=0.7, n_jobs=-1, cv=10 )
```

```python
import random


from eden.util import selection_iterator
def self_training(uri_pos, uri_neg, vectorizer=None, pos2neg_ratio=0.1, num_iterations=2,  threshold=0,  mode='less_than' ):
    n_jobs = -1
    #create pos data matrix
    from eden.converter.graph.gspan import gspan_to_eden
    graphs_pos = gspan_to_eden( uri_pos )
    Xpos = vectorizer.transform( graphs_pos, n_jobs=n_jobs )
    print 'Positives:'
    describe(Xpos)
    #select a fraction for the negatives
    num_pos = Xpos.shape[0]
    num_neg = int(float(num_pos) * pos2neg_ratio)
    #select the initial ids for the negatives as the first num_neg
    ids = range(num_neg)
    #iterate: select negatives and create a model using postives + selected negatives 
    for i in range(num_iterations):
        print 'Iteration: %d/%d'%(i+1,num_iterations)
        #select only a fraction of the negatives
        graphs_neg = gspan_to_eden( uri_neg )
        Xneg = vectorizer.transform( selection_iterator(graphs_neg,ids), n_jobs=n_jobs )
        print 'Negatives:'
        describe(Xneg)
        #fit the estimator on all positives and selected negatives
        from eden.util import fit_estimator
        estimator = fit_estimator( positive_data_matrix=Xpos, negative_data_matrix=Xneg, cv=10 )
        if i < num_iterations -1:
            #use the estimator to select the next batch of negatives
            graphs_neg = gspan_to_eden( uri_neg )
            ids = list()
            predictions = vectorizer.predict(graphs_neg, estimator)
            for i, prediction in enumerate(predictions):
                if mode == 'less_then':
                    comparison = prediction < float(threshold)
                else:
                    comparison = prediction > float(threshold)
                if comparison:
                    ids.append(i)
            #keep a random sample of num_neg difficult cases
            random.shuffle(ids)
            ids = ids[:num_neg]
            
    return estimator
```
