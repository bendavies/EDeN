```python
def performance_estimate(uri_pos, uri_neg, vectorizer, optpredictor=None):
    n_jobs = -1
    cv = 10
    from scipy.sparse import vstack
    import numpy as np
    from eden.converter.graph.gspan import gspan_to_eden
    graphs_pos = gspan_to_eden( uri_pos )
    Xpos = vectorizer.transform( graphs_pos, n_jobs=n_jobs )
    graphs_neg = gspan_to_eden( uri_neg )
    Xneg = vectorizer.transform( graphs_neg, n_jobs=n_jobs )
    yp =  [1] * Xpos.shape[0]
    yn = [-1] * Xneg.shape[0]
    y = np.array(yp + yn)
    X = vstack( [Xpos,Xneg] , format = "csr")
    describe(X)

    if optpredictor is None:
        from sklearn.linear_model import SGDClassifier
        from sklearn.grid_search import RandomizedSearchCV
        from sklearn import cross_validation
        from scipy.stats import randint
        from scipy.stats import uniform
        predictor = SGDClassifier(class_weight='auto', shuffle = True, n_jobs = n_jobs)
        #hyperparameter optimization
        param_dist = {"n_iter": randint(5, 100),
                      "power_t": uniform(0.1),
                      "alpha": uniform(1e-08,1e-03),
                      "eta0" : uniform(1e-03,10),
                      "penalty": ["l1", "l2", "elasticnet"],
                      "learning_rate": ["invscaling", "constant","optimal"]}
        scoring = 'roc_auc'
        n_iter_search = 20
        random_search = RandomizedSearchCV( predictor, param_distributions = param_dist, n_iter = n_iter_search, cv = cv, scoring = scoring, n_jobs = n_jobs )
        random_search.fit( X, y )
        optpredictor= SGDClassifier( class_weight='auto', shuffle = True, n_jobs = n_jobs, **random_search.best_params_ )
        #fit the predictor on all available data
        optpredictor.fit( X, y ) 
    
    
    from sklearn import cross_validation
    print 'Classifier:'
    print optpredictor
    print '-'*73

    print 'Predictive performance:'
    #assess the generalization capacity of the model via a 10-fold cross validation
    for scoring in ['accuracy','precision', 'recall', 'f1', 'average_precision', 'roc_auc']:
        scores = cross_validation.cross_val_score( optpredictor, X, y, cv = cv, scoring = scoring, n_jobs = n_jobs )
        print( '%20s: %.3f +- %.3f' % ( scoring, np.mean( scores ), np.std( scores ) ) )
    print '-'*73
```

```python
def selection_iterator(iterable, ids):
    '''Given an iterable and a list of ids (zero based) yield only the items whose id matches'''
    ids = sorted(ids)
    counter = 0
    for id,item in enumerate(iterable):
        if id == ids[counter]:
            yield item
            counter += 1
            if counter == len(ids):
                break
```

```python
import random

def describe(X):
    print 'Instances: %d ; Features: %d (with an avg of %d features per instance)' % (X.shape[0], X.shape[1],  X.getnnz()/X.shape[0])

            
def self_training(uri_pos, uri_neg, vectorizer=None, pos2neg_ratio=0.1, num_iterations=2,  threshold=0,  mode='less_than' ):
    n_jobs = -1
    #create pos data matrix
    from eden.converter.graph.gspan import gspan_to_eden
    graphs_pos = gspan_to_eden( uri_pos )
    Xpos = vectorizer.transform( graphs_pos, n_jobs=n_jobs )
    print 'Positives:'
    describe(Xpos)
    #select a fraction for the negatives
    num_pos = Xpos.shape[0]
    num_neg = int(float(num_pos) * pos2neg_ratio)
    #select the initial ids for the negatives as the first num_neg
    ids = range(num_neg)
    #iterate: select negatives and create a model using postives + selected negatives 
    for i in range(num_iterations):
        print 'Iteration: %d/%d'%(i+1,num_iterations)
        #select only a fraction of the negatives
        graphs_neg = gspan_to_eden( uri_neg )
        Xneg = vectorizer.transform( selection_iterator(graphs_neg,ids), n_jobs=n_jobs )
        print 'Negatives:'
        describe(Xneg)
        #fit the estimator on all positives and selected negatives
        from eden.util import fit_estimator
        estimator = fit_estimator( positive_data_matrix=Xpos, negative_data_matrix=Xneg, cv=10 )
        if i < num_iterations -1:
            #use the estimator to select the next batch of negatives
            graphs_neg = gspan_to_eden( uri_neg )
            ids = list()
            predictions = vectorizer.predict(graphs_neg, estimator)
            for i, prediction in enumerate(predictions):
                if mode == 'less_then':
                    comparison = prediction < float(threshold)
                else:
                    comparison = prediction > float(threshold)
                if comparison:
                    ids.append(i)
            #keep a random sample of num_neg difficult cases
            random.shuffle(ids)
            ids = ids[:num_neg]
            
    return estimator
```
